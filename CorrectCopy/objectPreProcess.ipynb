```python
idx=0
for V in videos: #os.listdir('.'):
    #if each_file.endswith("objects"):
    #    pass
    #if V == '.DS_Store' or V.size[0]/(V.size[0]+V.size[1]) < .45 or V.size[0] + V.size[1] < 350:
    print(V)
    image = Image.open(V)
    if image.size[0]/(image.size[0]+ image.size[1]) < .45 or image.size[0] + image.size[1] < 350:
        pass
    else:
        #111
        #image = Image.open(V)
        box = (0,  
               image.size[1] -  round(image.size[0]*.64, 0),
               image.size[0], 
               image.size[1] )
        cropped_image = image.crop(box)
        #cropped_image.show()
        #cropped_image.save('cropped_image.jpg')
        print(cropped_image.size, "::", V)
        new_image = cropped_image.resize((224,224) )
        #new_image.show()
        print(new_image.size)
        new_image.save(os.path.join(base_dir, 'X' + V) )
        idx += 1
        print(idx)
```
